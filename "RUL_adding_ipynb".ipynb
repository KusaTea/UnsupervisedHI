#RUL concating

```python
fd_003_train = pd.read_csv('clean_data.csv')
```

```python
fd_003_train
```

```python
fd_003_cycles = fd_003_train[["unit_number","time_cycles"]]
max_cycles = fd_003_cycles.groupby('unit_number').max()
max_cycles['unit_number'] = max_cycles.index
```

```python
fd_003_train['RUL'] = fd_003_train.groupby('unit_number')['time_cycles'].transform(max)
```

```python
for i in range(0,len(fd_003_train)):
  if fd_003_train['RUL'].loc[i] - fd_003_train['time_cycles'].loc[i] > 124:
    fd_003_train['RUL'].loc[i] = 125
  else:
    fd_003_train['RUL'].loc[i] = fd_003_train['RUL'].loc[i] - fd_003_train['time_cycles'].loc[i]
```
